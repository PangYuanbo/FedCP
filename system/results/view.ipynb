```python
import matplotlib.pyplot as plt

# 文件路径
file_path = 'results_mnist-0.1-npz_1000_0.0050.txt'

# 初始化数据存储
rounds = []
acc_values = []

# 读取文件并解析数据
with open(file_path, 'r') as file:
    lines = file.readlines()[1:]  # 跳过第一行
    for line in lines:
        parts = line.split()
        round_num = int(parts[1][:-1])  # 提取Round号，去掉冒号
        acc_value = float(parts[-1])   # 提取最后一个值作为ACC值
        rounds.append(round_num)
        acc_values.append(acc_value)

# 绘制折线图
plt.figure(figsize=(10, 6))
plt.plot(rounds, acc_values, label='Accuracy')  # 移除 marker='o'
plt.xlabel('Round')
plt.ylabel('Accuracy')
plt.title('Accuracy over Rounds')
plt.grid()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 文件路径
file_path1 = 'results_mnist-0.1-npz_200_0.0050_model.feature_extractor.txt'
file_path2 = 'results_mnist-0.1-npz_200_0.0050_model.head.txt'

# 初始化数据存储
rounds1, acc_values1 = [], []
rounds2, acc_values2 = [], []

# 读取文件1并解析数据
with open(file_path1, 'r') as file:
    lines = file.readlines()[1:]  # 跳过第一行
    for line in lines:
        parts = line.split()
        round_num = int(parts[1][:-1])  # 提取Round号，去掉冒号
        acc_value = float(parts[-1])   # 提取最后一个值作为ACC值
        rounds1.append(round_num)
        acc_values1.append(acc_value)

# 读取文件2并解析数据
with open(file_path2, 'r') as file:
    lines = file.readlines()[1:]  # 跳过第一行
    for line in lines:
        parts = line.split()
        round_num = int(parts[1][:-1])  # 提取Round号，去掉冒号
        acc_value = float(parts[-1])   # 提取最后一个值作为ACC值
        rounds2.append(round_num)
        acc_values2.append(acc_value)

# 绘制合并图
plt.figure(figsize=(10, 6))
plt.plot(rounds1, acc_values1, label='File 1 Accuracy')
plt.plot(rounds2, acc_values2, label='File 2 Accuracy')
plt.xlabel('Round')
plt.ylabel('Accuracy')
plt.title('Accuracy over Rounds (Combined)')
plt.grid()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 文件路径
file_path = '11.txt'

# 初始化数据存储
rounds_file1, acc_values_file1 = [], []
rounds_file2, acc_values_file2 = [], []

# 读取文件并解析数据
with open(file_path, 'r') as file:
    lines = file.readlines()
    for line in lines:
        parts = line.split()
        round_num = int(parts[1][:-1])  # 提取Round号，去掉冒号
        acc_value = float(parts[-1])   # 提取ACC值

        # 判断属于哪个文件的数据
        if round_num in rounds_file1:  # 如果Round重复，归类到第二组
            rounds_file2.append(round_num)
            acc_values_file2.append(acc_value)
        else:
            rounds_file1.append(round_num)
            acc_values_file1.append(acc_value)

# 绘制合并图
plt.figure(figsize=(10, 6))
plt.plot(rounds_file1, acc_values_file1, label='File 1 Accuracy')
plt.plot(rounds_file2, acc_values_file2, label='File 2 Accuracy')
plt.xlabel('Round')
plt.ylabel('Accuracy')
plt.title('Accuracy over Rounds (Separated)')
plt.grid()
plt.legend()
plt.show()
```
